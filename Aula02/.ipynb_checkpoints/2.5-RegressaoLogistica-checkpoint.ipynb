### Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Leitura dos dados de diabets

```python
df = pd.read_csv('diabetes.csv')
```

```python
df.head()
```

### Informações estatísticas dos dados 

```python
df.describe()
```

### Gerando gráfico

```python
plt.plot(df['Age'], df['Pregnancies'], 'o')
plt.xlabel('Idade')
plt.ylabel('Gravidez')
plt.show()
```

### Separação de variável dependente 

```python
y = df['Outcome']
```

```python
X = df.drop('Outcome', axis = 1)      #df menos Outcome
X.head()
```

### Separando dados em treino e teste

```python
from sklearn.model_selection import train_test_split
```

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size = 0.2)
```

### Selecionar modelo de regressão 

```python
from sklearn.linear_model import LogisticRegression
```

```python
modelo = LogisticRegression(max_iter = 5000)
```

### Treinamento do modelo

```python
modelo.fit(X_treino, y_treino)
```

### Realizando previsões

```python
y_previsto = modelo.predict(X_teste)
```

```python
print(y_previsto[:10])
```

### Validando modelo

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_teste, y_previsto)
```

```python
print(accuracy * 100)     #Capaz de prever 75%
```
